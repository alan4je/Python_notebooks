```python
import pandas as pd
import urllib
import json
```

```python
# Importing 311 data from Berkeley Open Data Portal
endpoint_url = 'https://data.cityofberkeley.info/resource/k489-uv4i.json'
# open a connection to the URL
connection = urllib.urlopen(endpoint_url)

# download the results
results = connection.read()

# parse the string into a Python data structure
data = json.loads(results)
```

```python
# turn the json data into a dataframe
df = pd.DataFrame(data)
print len(df)
df.head()
```

```python
df = df[df.latitude.isnull() == False]
print len(df)
```

# Performing cluster analysis

```python
# magic command to display matplotlib plots inline within the ipython notebook
%matplotlib inline

# import necessary modules
import pandas as pd, numpy as np, matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn import metrics
# from geopy.distance import great_circle
from shapely.geometry import MultiPoint
```

```python
df.describe()
```

```python
df.latitude = df.latitude.astype('float64')
df.longitude = df.longitude.astype('float64')
```

```python
# Getting a sense of how the data looks like
df.plot(kind='scatter', x='longitude', y='latitude', figsize=[10,7])
```

```python
# convert the lat/long columns from dataframe to a numpy matrix
df = df.sort('longitude', ascending=False)
coordinates = df.as_matrix(columns=['longitude', 'latitude'])
coordinates
```

```python
# eps is the physical distance from each point that forms its -neighborhood
# min_samples is the minimum cluster size, otherwise the point(s) are noise
db = DBSCAN(eps=0.0025, min_samples=5).fit(coordinates)
cluster_labels = db.labels_
unique_labels = set(cluster_labels)

# calculate the number of clusters in labels, ignoring noise (which is labeled -1) if present
num_clusters = len(set(cluster_labels)) - (1 if -1 in cluster_labels else 0)

str(list(cluster_labels))
```

```python
# get colors and plot all the points, color-coded by cluster (or gray if not in any cluster, aka noise)
fig, ax = plt.subplots(figsize=[10,7])
colors = plt.cm.rainbow(np.linspace(0, 1, len(unique_labels)))

# for each cluster label and color, plot the cluster's points
for cluster_label, color in zip(unique_labels, colors):
    
    size = 150
    if cluster_label == -1: #make the noise (which is labeled -1) appear as smaller gray points
        color = 'gray'
        size = 30
    
    # plot the points that match the current cluster label
    x_coords = coordinates[cluster_labels==cluster_label][:,0]
    y_coords = coordinates[cluster_labels==cluster_label][:,1]
    ax.scatter(x=x_coords, y=y_coords, c=color, edgecolor='k', s=size, alpha=0.5)

ax.set_title('Number of clusters: {}'.format(num_clusters))
plt.show()
```

In the analysis above, I performed clusted analysis on 311 calls in Berkeley. Data was obtained the Berkeley Open Data Portal. In the analysis, I used an epsilon value of 0.0025 in order to distribute the data points into a significant amount of clusters. I kept changing the values of epsilon and min_sampe variables till I obtained at-least 20 clusters. The minimum sample size of 5 in such a small radius is a good enough indicator for high concentration of calls from the specific zone, and probably worth noticing. The sections with less than 5 calls were not highlighted, and could potentially be safer or lower-risk areas.
